```python
def percent_diff(value1,err1,value2,err2):
    if value1 < value2 :
        delta1 = err1[1]
        delta2 = err2[0]
    else :
        delta1 = err1[0]
        delta2 = err2[1]
    Uncertainty = ((delta1)**2 + (delta2)**2)**0.5

    return np.round(np.abs(value1-value2)/), Uncertainty

```
